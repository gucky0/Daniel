```python
#importing the time module
import sys, csv, os, random
import pandas as pd
import matplotlib.pyplot as plt

dct = {}
fields=['username', 'score']
filename = "leaderboard.csv"

wordDict = {'Colors':'red orange yellow green blue indigo violet white black brown'.split(), \
    'Shapes':'square triangle rectangle circle ellipse rhombus trapezoid chevron pentagon hexagon septagon octagon'.split(), \
    'Fruits':'apple orange lemon lime pear watermelon grape grapefruit cherry banana cantaloupe mango strawberry tomato'.split(), \
    'Animals':'bat bear beaver cat cougar crab deer dog donkey duck eagle fish frog goat leech lion lizard monkey moose mouse otter owl panda python rabbit rat shark sheep skunk squid tiger turkey turtle weasel whale wolf wombat zebra'.split()}

keyList = list(wordDict.keys())

# get fisrt letter from keys
firstKeys = [_[0].lower() for _ in keyList]

choiceDict = dict(zip(firstKeys, keyList))

random_key = random.choice(keyList)
choiceDict['r'] = 'random'

prompt = f'''please choose a word from this list: {choiceDict}: '''

def betterInput(prompt):
    res = input(prompt)
    if res == 'quit':
        # sys.exit("Goodbye!")
        raise SystemExit("Goodbye!")
    return res

def getWord():
    wordKey = None
    while wordKey == None:
        userInput = betterInput(prompt)
        choiceDict['r'] = random_key
        wordKey = choiceDict.get(userInput) #input(f"please choose a word from this list or random ('r'): {keyList}")
    word = random.choice(wordDict.get(wordKey))
    print(f"This game is about: {wordKey}")
    return word

def updateLeaderboard(username, score):
    global dct

    #create file if it doesnt exist
    if not os.path.exists(filename):
        open(filename, 'w').close()
    else:
        #read data from csv
        with open (filename, "r", newline = '') as file:  
            reader=csv.reader(file)
            next(reader)
            for line in reader:
                #sort for username and score
                key = line[0]
                value = int(line[1])
                dct[key] = value
    
    #create / update data for current match
    if username in dct:
        if score > dct[username]:
            dct[username] = score
    else:
        dct[username] = score

    #sort data based on highest to lowest
    dct = dict(sorted(dct.items(), key=lambda _: _[1], reverse=True))

    with open (filename, "w", newline = '') as file:
        header = ['username', 'score']
        headerwriter=csv.writer(file)
        headerwriter.writerow(header)
        writer=csv.DictWriter(file, fieldnames=fields)
        for username,score in dct.items():
            writer.writerow({'username' : username, 'score' : int(score)})

def displayLeaderboard():
    data = pd.read_csv(filename)
    data.head()
    df = pd.DataFrame(data)
    x = df['username']
    y = df['score']

    fig = plt.figure(figsize = (10, 5))
    ax = fig.add_axes([0,0,1,1])
    ax.set_ylabel('Scores',fontsize = 16)
    ax.set_title('Hangman Leaderboard',fontsize = 32)
    for i in range(len(df)):
        plt.text(i-0.1, 50, y.values[i], fontsize = 20, fontweight ='bold', color ='black')

    plt.bar(x,y, color ='maroon')
    plt.show()

def hangman():
    username = betterInput("What is your name? ")
    print ("Hello,", username + "! Time to play hangman!\n")

    word = getWord()

    #implement 60 second rule
    # print ("Start guessing...")

    #creates a variable with an empty value
    guesses = ''
    rejected = []

    #determine the number of turns
    turns = 10
    # Create a while loop
    #check if the turns are more than zero
    while turns > 0:         
        # make a counter that starts with zero
        failed = 0             
        # for every character in secret_word    
        for char in word:      
        # see if the character is in the players guess
            if char in guesses:    
                # print then out the character
                print (char, end = ' ') 
            elif char == ' ':
                print(' ', end = '')
            else:
                # if not found, print a dash
                print ("_", end = ' ')    
                # and increase the failed counter with one
                failed += 1    
        print('\n')
        # if failed is equal to zero
        # print You Won
        if failed == 0:
            # print(username, turns*10)
            updateLeaderboard(username, turns*10)   
            displayLeaderboard()     
            print  ("You won")  
            choice = betterInput("Would you like to play again? y/n\n")
            if "y" in choice:
                hangman()
            elif "n" in choice:
                sys.exit
                break
            else:
                print("Something went wrong, type y or n.")   

        # ask the user go guess a character
        guess = '1'
        while guess in rejected or not guess.isalpha() or len(guess)>1:
            # user cannot guess numbers or previous incorrect guesses
            guess = betterInput("guess a character: ")
        # set the players guess to guesses
        guesses += guess                    
        # if the guess is not found in the secret word
        if guess not in word:  
            # turns counter decreases with 1 (now 9)
            turns -= 1        
            # print wrong
            print ("\nWrong")   
            # how many turns are left
            rejected.append(guess)
            rejected = list(set(rejected))
            print(f"Rejected letters: {rejected}")
            print ("You have", + turns, 'more guesses' )
            # if the turns are equal to zero
            if turns == 0:           
            # print "You Lose"
                print ("You Lose")



hangman()
```
